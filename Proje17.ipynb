```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QLabel, QLineEdit, QPushButton, QTextEdit

class Ders:
    def _init_(self, adi, icerik, ogretmen):
        self.adi = adi
        self.icerik = icerik
        self.ogretmen = ogretmen
        self.materyaller = []

    def materyal_yukle(self, materyal):
        self.materyaller.append(materyal)

    def materyal_eris(self):
        return self.materyaller

class Ogrenci:
     def _init_(self, adi, soyadi, numara):
        self.adi = adi
        self.soyadi = soyadi
        self.numara = numara
        self.katildigi_dersler = []

class Materyal:
    def _init_(self, adi, turu, icerik):
        self.adi = adi
        self.turu = turu
        self.icerik = icerik

class Arayuz(QWidget):
    def _init_(self):
        super()._init_()

        self.init_ui()

    def init_ui(self):
        self.setWindowTitle('Ders ve Materyal Yönetim Arayüzü')

        self.ogrenci_adi_label = QLabel('Öğrenci Adı:')
        self.ogrenci_adi_input = QLineEdit()

        self.ogrenci_soyadi_label = QLabel('Öğrenci Soyadı:')
        self.ogrenci_soyadi_input = QLineEdit()

        self.ogrenci_numara_label = QLabel('Öğrenci Numarası:')
        self.ogrenci_numara_input = QLineEdit()

        self.ders_adi_label = QLabel('Ders Adı:')
        self.ders_adi_input = QLineEdit()

        self.materyal_adi_label = QLabel('Materyal Adı:')
        self.materyal_adi_input = QLineEdit()

        self.ekle_ogrenci_button = QPushButton('Öğrenci Ekle')
        self.ekle_ogrenci_button.clicked.connect(self.ogrenci_ekle)

        self.ekle_ders_button = QPushButton('Ders Ekle')
        self.ekle_ders_button.clicked.connect(self.ders_ekle)

        self.ekle_materyal_button = QPushButton('Materyal Ekle')
        self.ekle_materyal_button.clicked.connect(self.materyal_ekle)

        self.bilgi_textbox = QTextEdit()

        self.layout = QVBoxLayout()
        self.layout.addWidget(self.ogrenci_adi_label)
        self.layout.addWidget(self.ogrenci_adi_input)
        self.layout.addWidget(self.ogrenci_soyadi_label)
        self.layout.addWidget(self.ogrenci_soyadi_input)
        self.layout.addWidget(self.ogrenci_numara_label)
        self.layout.addWidget(self.ogrenci_numara_input)
        self.layout.addWidget(self.ekle_ogrenci_button)
        self.layout.addWidget(self.ders_adi_label)
        self.layout.addWidget(self.ders_adi_input)
        self.layout.addWidget(self.ekle_ders_button)
        self.layout.addWidget(self.materyal_adi_label)
        self.layout.addWidget(self.materyal_adi_input)
        self.layout.addWidget(self.ekle_materyal_button)
        self.layout.addWidget(self.bilgi_textbox)

        self.setLayout(self.layout)

    def ogrenci_ekle(self):
        adi = self.ogrenci_adi_input.text()
        soyadi = self.ogrenci_soyadi_input.text()
        numara = self.ogrenci_numara_input.text()

        self.ogrenci = Ogrenci(adi, soyadi, numara)
        self.bilgi_textbox.append(f"Yeni öğrenci eklendi: {adi} {soyadi} ({numara})")

    def ders_ekle(self):
        adi = self.ders_adi_input.text()
        self.ders = Ders(adi, '', '')
        self.ogrenci.katildigi_dersler.append(self.ders)
        self.bilgi_textbox.append(f"Yeni ders eklendi: {adi}")

    def materyal_ekle(self):
        adi = self.materyal_adi_input.text()
        self.materyal = Materyal(adi, '', '')
        self.ders.materyal_yukle(self.materyal)
        self.bilgi_textbox.append(f"Yeni materyal eklendi: {adi}")

if _name_ == '_main_':
    app = QApplication(sys.argv)
    arayuz = Arayuz()
    arayuz.show()
    sys.exit(app.exec_())
```
