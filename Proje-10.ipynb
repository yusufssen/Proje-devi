```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QLabel, QLineEdit, QPushButton, QTextEdit

class MusteriUygulamasi(QWidget):
    def _init_(self):
        super()._init_()
        self.setWindowTitle("Müşteri İlişkileri Yönetimi (CRM)")
        self.musteri = Musteri("Ahmet Yılmaz", "ahmet@email.com, 555-555-5555")
        self.init_ui()

    def init_ui(self):
        # Müşteri Bilgileri Görüntüleme
        self.musteri_bilgileri_label = QLabel("Müşteri Bilgileri:")
        self.musteri_bilgileri_text = QTextEdit(str(self.musteri))
        self.musteri_bilgileri_text.setReadOnly(True)

        # Satış Takibi Yapma
        self.satis_numarasi_label = QLabel("Satış Numarası:")
        self.satis_numarasi_input = QLineEdit()
        self.urunler_label = QLabel("Satılan Ürünler:")
        self.urunler_input = QLineEdit()
        self.satis_takibi_button = QPushButton("Satış Ekle")
        self.satis_takibi_button.clicked.connect(self.satis_takibi_clicked)

        # Destek Taleplerini Yönetme
        self.talep_numarasi_label = QLabel("Talep Numarası:")
        self.talep_numarasi_input = QLineEdit()
        self.detaylar_label = QLabel("Detaylar:")
        self.detaylar_input = QLineEdit()
        self.destek_talebi_button = QPushButton("Destek Talebi Oluştur")
        self.destek_talebi_button.clicked.connect(self.destek_talebi_clicked)

        # Ana düzen
        main_layout = QVBoxLayout()
        main_layout.addWidget(self.musteri_bilgileri_label)
        main_layout.addWidget(self.musteri_bilgileri_text)
        main_layout.addWidget(self.satis_numarasi_label)
        main_layout.addWidget(self.satis_numarasi_input)
        main_layout.addWidget(self.urunler_label)
        main_layout.addWidget(self.urunler_input)
        main_layout.addWidget(self.satis_takibi_button)
        main_layout.addWidget(self.talep_numarasi_label)
        main_layout.addWidget(self.talep_numarasi_input)
        main_layout.addWidget(self.detaylar_label)
        main_layout.addWidget(self.detaylar_input)
        main_layout.addWidget(self.destek_talebi_button)

        self.setLayout(main_layout)

    def satis_takibi_clicked(self):
        satis_numarasi = self.satis_numarasi_input.text()
        urunler = self.urunler_input.text().split(',')
        self.musteri.satış_ekle(satis_numarasi, urunler)
        self.musteri_bilgileri_text.setPlainText(str(self.musteri))

    def destek_talebi_clicked(self):
        talep_numarasi = self.talep_numarasi_input.text()
        detaylar = self.detaylar_input.text()
        self.musteri.destek_talebi_oluştur(talep_numarasi, detaylar)
        self.musteri_bilgileri_text.setPlainText(str(self.musteri))


if _name_ == "_main_":
    app = QApplication(sys.argv)
    window = MusteriUygulamasi()
    window.show()
    sys.exit(app.exec_())
```
