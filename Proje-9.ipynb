```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QLabel, QLineEdit, QPushButton, QTextEdit

class SeyahatUygulamasi(QWidget):
    def _init_(self):
        super()._init_()
        self.setWindowTitle("Seyahat Uygulaması")
        self.seyahat = Seyahat()  # Seyahat sınıfından bir örnek oluşturuyoruz
        self.init_ui()

    def init_ui(self):
        # Rota Ekleme
        self.rota_baslangic_label = QLabel("Başlangıç:")
        self.rota_baslangic_input = QLineEdit()
        self.rota_bitis_label = QLabel("Bitiş:")
        self.rota_bitis_input = QLineEdit()
        self.rota_uzaklik_label = QLabel("Uzaklık (km):")
        self.rota_uzaklik_input = QLineEdit()
        self.rota_ekle_button = QPushButton("Rota Ekle")
        self.rota_ekle_button.clicked.connect(self.rota_ekle_clicked)

        rota_layout = QVBoxLayout()
        rota_layout.addWidget(self.rota_baslangic_label)
        rota_layout.addWidget(self.rota_baslangic_input)
        rota_layout.addWidget(self.rota_bitis_label)
        rota_layout.addWidget(self.rota_bitis_input)
        rota_layout.addWidget(self.rota_uzaklik_label)
        rota_layout.addWidget(self.rota_uzaklik_input)
        rota_layout.addWidget(self.rota_ekle_button)

        # Konaklama Seçme
        self.otel_adi_label = QLabel("Otel Adı:")
        self.otel_adi_input = QLineEdit()
        self.konum_label = QLabel("Konum:")
        self.konum_input = QLineEdit()
        self.fiyat_label = QLabel("Fiyat ($/gece):")
        self.fiyat_input = QLineEdit()
        self.konaklama_sec_button = QPushButton("Konaklama Seç")
        self.konaklama_sec_button.clicked.connect(self.konaklama_sec_clicked)

        konaklama_layout = QVBoxLayout()
        konaklama_layout.addWidget(self.otel_adi_label)
        konaklama_layout.addWidget(self.otel_adi_input)
        konaklama_layout.addWidget(self.konum_label)
        konaklama_layout.addWidget(self.konum_input)
        konaklama_layout.addWidget(self.fiyat_label)
        konaklama_layout.addWidget(self.fiyat_input)
        konaklama_layout.addWidget(self.konaklama_sec_button)

        # Seyahat Planı Düzenleme
        self.rota_index_label = QLabel("Rota İndeksi:")
        self.rota_index_input = QLineEdit()
        self.konaklama_index_label = QLabel("Konaklama İndeksi:")
        self.konaklama_index_input = QLineEdit()
        self.plan_duzenle_button = QPushButton("Seyahat Planı Düzenle")
        self.plan_duzenle_button.clicked.connect(self.plan_duzenle_clicked)

        plan_layout = QVBoxLayout()
        plan_layout.addWidget(self.rota_index_label)
        plan_layout.addWidget(self.rota_index_input)
        plan_layout.addWidget(self.konaklama_index_label)
        plan_layout.addWidget(self.konaklama_index_input)
        plan_layout.addWidget(self.plan_duzenle_button)

        # Sonuçları Gösterme
        self.sonuclar_label = QLabel("Sonuçlar:")
        self.sonuclar_text = QTextEdit()

        sonuclar_layout = QVBoxLayout()
        sonuclar_layout.addWidget(self.sonuclar_label)
        sonuclar_layout.addWidget(self.sonuclar_text)

        # Ana düzen
        main_layout = QHBoxLayout()
        main_layout.addLayout(rota_layout)
        main_layout.addLayout(konaklama_layout)
        main_layout.addLayout(plan_layout)
        main_layout.addLayout(sonuclar_layout)

        self.setLayout(main_layout)

    def rota_ekle_clicked(self):
        baslangic = self.rota_baslangic_input.text()
        bitis = self.rota_bitis_input.text()
        uzaklik = int(self.rota_uzaklik_input.text())
        self.seyahat.rota_ekle(baslangic, bitis, uzaklik)
        self.sonuclar_text.append("Rota eklendi: {} - {}, Uzaklık: {} km".format(baslangic, bitis, uzaklik))

    def konaklama_sec_clicked(self):
        otel_adi = self.otel_adi_input.text()
        konum = self.konum_input.text()
        fiyat = float(self.fiyat_input.text())
        self.seyahat.konaklama_sec(otel_adi, konum, fiyat)
        self.sonuclar_text.append("Konaklama seçildi: {}, Konum: {}, Fiyat: ${}/gece".format(otel_adi, konum, fiyat))

    def plan_duzenle_clicked(self):
        rota_index = int(self.rota_index_input.text())
        konaklama_index = int(self.konaklama_index_input.text())
        self.seyahat.plan_duzenle(rota_index, konaklama_index)
        seyahat_planı = self.seyahat.seyahat_planlari[-1]
        rota = seyahat_planı.rota
        konaklama = seyahat_planı.konaklama
        self.sonuclar_text.append("Seyahat Planı:")
        self.sonuclar_text.append("- Rota: {} - {}, Uzaklık: {} km".format(rota.baslangic, rota.bitis, rota.uzaklik))
        self.sonuclar_text.append("- Konaklama: {}, Konum: {}, Fiyat: ${}/gece".format(konaklama.otel_adi, konaklama.konum, konaklama.fiyat))

if _name_ == "_main_":
    app = QApplication(sys.argv)
    window = SeyahatUygulamasi()
    window.show()
    sys.exit(app.exec_())
```
