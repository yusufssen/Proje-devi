```python
mport sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QLabel, QLineEdit, QPushButton, QTextEdit

class Sporcu:
    def _init_(self, ad):
        self.ad = ad
        self.antrenmanlar = {}

    def program_olustur(self, antrenman_adi, detaylar):
        self.antrenmanlar[antrenman_adi] = detaylar

    def antreman_takibi(self):
        takip = f"{self.ad}'ın antrenman programı:\n"
        for antrenman, detaylar in self.antrenmanlar.items():
            takip += f"{antrenman}: {detaylar}\n"
        return takip

class SporcuEkleWidget(QWidget):
    def _init_(self, sporcular):
        super()._init_()
        self.sporcular = sporcular
        self.init_ui()

    def init_ui(self):
        self.setWindowTitle("Sporcu Ekle")

        self.ad_label = QLabel("Sporcu Adı:")
        self.ad_input = QLineEdit()
        self.ekle_button = QPushButton("Sporcu Ekle")
        self.ekle_button.clicked.connect(self.sporcu_ekle)

        v_box = QVBoxLayout()
        v_box.addWidget(self.ad_label)
        v_box.addWidget(self.ad_input)
        v_box.addWidget(self.ekle_button)

        self.setLayout(v_box)

    def sporcu_ekle(self):
        ad = self.ad_input.text()
        self.sporcular[ad] = Sporcu(ad)
        print(f"{ad} sporcu olarak eklendi.")

class SporcuProgramWidget(QWidget):
    def _init_(self, sporcular, antrenmanlar):
        super()._init_()
        self.sporcular = sporcular
        self.antrenmanlar = antrenmanlar
        self.init_ui()

    def init_ui(self):
        self.setWindowTitle("Antrenman Programı Oluştur")

        self.sporcu_label = QLabel("Sporcu Adı:")
        self.sporcu_input = QLineEdit()
        self.antrenman_label = QLabel("Antrenman Adı:")
        self.antrenman_input = QLineEdit()
        self.ekle_button = QPushButton("Program Oluştur")
        self.ekle_button.clicked.connect(self.program_olustur)

        v_box = QVBoxLayout()
        v_box.addWidget(self.sporcu_label)
        v_box.addWidget(self.sporcu_input)
        v_box.addWidget(self.antrenman_label)
        v_box.addWidget(self.antrenman_input)
        v_box.addWidget(self.ekle_button)

        self.setLayout(v_box)

    def program_olustur(self):
        sporcu_ad = self.sporcu_input.text()
        antrenman_ad = self.antrenman_input.text()

        if sporcu_ad in self.sporcular and antrenman_ad in self.antrenmanlar:
            self.sporcular[sporcu_ad].program_olustur(antrenman_ad, self.antrenmanlar[antrenman_ad])
            print(f"{sporcu_ad} için {antrenman_ad} antrenman programı oluşturuldu.")
        else:
            print("Sporcu veya antrenman bulunamadı.")

class AntrenmanTakipWidget(QWidget):
    def _init_(self, sporcular):
        super()._init_()
        self.sporcular = sporcular
        self.init_ui()

    def init_ui(self):
        self.setWindowTitle("Antrenman Takibi")

        self.sporcu_label = QLabel("Sporcu Adı:")
        self.sporcu_input = QLineEdit()
        self.takip_label = QLabel()
        self.takip_button = QPushButton("Takibi Göster")
        self.takip_button.clicked.connect(self.takibi_goster)

        v_box = QVBoxLayout()
        v_box.addWidget(self.sporcu_label)
        v_box.addWidget(self.sporcu_input)
        v_box.addWidget(self.takip_button)
        v_box.addWidget(self.takip_label)

        self.setLayout(v_box)

    def takibi_goster(self):
        sporcu_ad = self.sporcu_input.text()
        if sporcu_ad in self.sporcular:
            takip = self.sporcular[sporcu_ad].antreman_takibi()
            self.takip_label.setText(takip)
        else:
            self.takip_label.setText("Sporcu bulunamadı.")

class SporcuUygulamasi(QWidget):
    def _init_(self):
        super()._init_()
        self.sporcular = {}
        self.antrenmanlar = {"Kosu": "30 dakika koşu", "Kardiyo": "1 saat bisiklet"}

        self.setWindowTitle("Sporcu Uygulaması")
        self.layout = QVBoxLayout()

        self.sporcu_ekle_widget = SporcuEkleWidget(self.sporcular)
        self.layout.addWidget(self.sporcu_ekle_widget)

        self.program_widget = SporcuProgramWidget(self.sporcular, self.antrenmanlar)
        self.layout.addWidget(self.program_widget)

        self.antrenman_takip_widget = AntrenmanTakipWidget(self.sporcular)
        self.layout.addWidget(self.antrenman_takip_widget)

        self.setLayout(self.layout)
        self.show()

if _name_ == "_main_":
    app = QApplication(sys.argv)
    uygulama = SporcuUygulamasi()
    sys.exit(app.exec_())
```
