```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QLabel, QLineEdit, QPushButton, QTextEdit

class SiparisYonetimArayuzu(QWidget):
    def _init_(self):
        super()._init_()

        self.init_ui()

    def init_ui(self):
        self.setWindowTitle("Sipariş Yönetim Arayüzü")

        self.siparis_no_label = QLabel("Sipariş Numarası:")
        self.siparis_no_edit = QLineEdit()

        self.ad_label = QLabel("Müşteri Adı:")
        self.ad_edit = QLineEdit()

        self.adres_label = QLabel("Müşteri Adresi:")
        self.adres_edit = QLineEdit()

        self.urun_label = QLabel("Ürün Adı:")
        self.urun_edit = QLineEdit()

        self.miktar_label = QLabel("Miktar:")
        self.miktar_edit = QLineEdit()

        self.siparis_ekle_button = QPushButton("Sipariş Ekle")
        self.siparis_ekle_button.clicked.connect(self.siparis_ekle)

        self.siparisler_textedit = QTextEdit()

        vbox = QVBoxLayout()
        hbox1 = QHBoxLayout()
        hbox1.addWidget(self.siparis_no_label)
        hbox1.addWidget(self.siparis_no_edit)
        hbox1.addWidget(self.ad_label)
        hbox1.addWidget(self.ad_edit)
        hbox1.addWidget(self.adres_label)
        hbox1.addWidget(self.adres_edit)

        hbox2 = QHBoxLayout()
        hbox2.addWidget(self.urun_label)
        hbox2.addWidget(self.urun_edit)
        hbox2.addWidget(self.miktar_label)
        hbox2.addWidget(self.miktar_edit)
        hbox2.addWidget(self.siparis_ekle_button)

        vbox.addLayout(hbox1)
        vbox.addLayout(hbox2)
        vbox.addWidget(self.siparisler_textedit)

        self.setLayout(vbox)

    def siparis_ekle(self):
        siparis_no = self.siparis_no_edit.text()
        ad = self.ad_edit.text()
        adres = self.adres_edit.text()
        urun = self.urun_edit.text()
        miktar = self.miktar_edit.text()

        siparis_str = f"Sipariş Numarası: {siparis_no}\nMüşteri Adı: {ad}\nMüşteri Adresi: {adres}\nÜrün Adı: {urun}\nMiktar: {miktar}\n\n"
        self.siparisler_textedit.append(siparis_str)

if _name_ == '_main_':
    app = QApplication(sys.argv)
    arayuz = SiparisYonetimArayuzu()
    arayuz.show()
    sys.exit(app.exec_())
```
